# Players with more wage

```python
salary= df[['short_name','wage_eur']]
```

```python
salary
```

```python
player_salary=salary.sort_values(by=['wage_eur'],ascending=False)[0:5]
```

```python
player_salary
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5],list(player_salary['wage_eur'])[0:5],color=["Red","Green","blue","black","orange"])
plt.show()
```
